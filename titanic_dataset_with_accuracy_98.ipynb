```python
#making traing nad test datset
x_train= dataset.iloc[:,[2,4,5,6,7,9]]
y_train= dataset.iloc[:,1:2]
x_test = dataset1.iloc[:,[1,3,4,5,6,8]]
y_test = dataset3.iloc[:,1:2]
```

```python
x_train1=x_train.values
x_test1=x_test.values
y_train1=y_train.values
y_test1=y_test.values
```

```python
from sklearn.preprocessing import Imputer
imputer=Imputer(missing_values='NaN',strategy='mean',axis=0)
imputer.fit(x_train1[:,[2]])
x_train1[:,[2]]=imputer.transform(x_train1[:,[2]])
display(x_train1)
imputer=Imputer(missing_values='NaN',strategy='mean',axis=0)
imputer.fit(x_test1[:,[2,5]])
x_test1[:,[2,5]]=imputer.transform(x_test1[:,[2,5]])
display(x_test1)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder_x=LabelEncoder()
x_train1[:,1]=labelencoder_x.fit_transform(x_train1[:,1])
display(x_train1)
labelencoder_x1=LabelEncoder()
x_test1[:,1]=labelencoder_x1.fit_transform(x_test1[:,1])
display(x_test1)
```

```python
x_train1.shape,x_test1.shape
```

```python
onehotencoder = OneHotEncoder(categorical_features=[1])
x_train1 = onehotencoder.fit_transform(x_train1).toarray()
display(x_train1)
onehotencoder = OneHotEncoder(categorical_features=[1])
x_test1 = onehotencoder.fit_transform(x_test1).toarray()
display(x_test1)
```

```python
x_train1.shape,x_test1.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
x_train1 = sc_X.fit_transform(x_train1)
x_test1 = sc_X.fit_transform(x_test1)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=None)
x_train1=pca.fit_transform(x_train1)
x_test1=pca.transform(x_test1)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(C=0.01)
lr.fit(x_train1,y_train1)
pred=lr.predict(x_test1)
pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.metrics import f1_score,recall_score,precision_score
```

```python
accuracy_score(pred,y_test1)
```

```python
f1_score(pred,y_test1)
```

```python
recall_score(pred,y_test1)
```

```python
precision_score(pred,y_test1)
```
